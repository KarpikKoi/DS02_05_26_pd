***Zadanie 9  Confusion matrix i metryki***

Masz confusion matrix dla klasyfikacji binarnej: \
cm = [\
[80, 10], # TN, FP\
[5, 45] # FN, TP\
]\
Wyciągnij wartości TN, FP, FN, TP przez dwuwymiarowe indeksowanie i oblicz:\
accuracy = (TP + TN) / (TP + TN + FP + FN) ,\
precision = TP / (TP + FP) ,\
recall = TP / (TP + FN) .\
Wyświetl wyniki w formacie procentowym z dwoma miejscami po przecinku.

```python
cm = [
    [80,10], #TN,FP
    [5,45] #FN, TP
]

TP = cm[1][1]
TN = cm[0][0]
FP = cm[0][1]
FN = cm[1][0]

accuracy = round((TP + TN)/(TP+TN+FP+FN),4)
precision = round(TP / (TP + FP),4)
recall = round(TP / (TP + FN),4)

accuracy_procent = str(accuracy*100)+'%'
precision_procent = str(round((precision*100),2))+'%'
recall_procent = str(recall*100)+'%'

print(accuracy_procent)
print(precision_procent)
print(recall_procent)
```
